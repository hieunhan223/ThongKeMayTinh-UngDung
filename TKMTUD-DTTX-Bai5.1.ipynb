## <a name="gioi_thieu"/>Giới thiệu

**Trực quan hóa dữ liệu** (data visualization) là công đoạn rất quan trọng trong phân tích dữ liệu với mục đích chính là **khám phá** (explore) và **mô tả** (describe) dữ liệu. Đây là phần quan trọng trong **phân tích thăm dò dữ liệu** (Exploratory Data Analysis) hay **thống kê mô tả** (Descriptive Statistics).

[**Matplotlib**](https://matplotlib.org/) là thư viện trực quan hóa phổ biến trong Python. Nhiều thư viện đồ họa khác, như [**seaborn**](https://seaborn.pydata.org/), được xây dựng dựa trên Matplotlib.

Tham khảo: [Getting started](https://matplotlib.org/stable/tutorials/introductory/quick_start.html#sphx-glr-tutorials-introductory-quick-start-py), [Tutorial](https://matplotlib.org/stable/tutorials/index.html), [User guide](https://matplotlib.org/stable/users/index), [Reference](https://matplotlib.org/stable/api/index.html).

Nạp các thư viện với tên qui ước.

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

Matplotlib có thể được dùng theo 2 cách.

Cách thứ nhất là **dựa trên trạng thái** (state-based) tương tự như MATLAB.

```python
x = np.linspace(0, 4*np.pi, 100)

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

plt.xlabel("x")
plt.ylabel("cos(x)")

plt.show()
```

Trong cách này, khi ta dùng các hàm trong `plt`, matplotlib ghi nhớ biểu đồ hiện tại và vẽ lên biểu đồ đó.

Cách thứ hai, linh hoạt hơn, là **hướng đối tượng** (object-oriented).

```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))

ax[1].set_xlabel("x")
ax[0].set_ylabel("sin(x)")
ax[1].set_ylabel("cos(x)")

plt.show()
```

Trong cách này, ta dùng các tham chiếu đến các biểu đồ (axes) và gọi các phương thức trên đó để trực tiếp vẽ lên biểu đồ tương ứng.

Ta có thể đặt style cho các biểu đồ.

```python
plt.style.use("dark_background")
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.legend(["sin(x)", "cos(x)"])
plt.show()
```

Matplotlib hỗ trợ nhiều style khác nhau. Tham khảo: [Style sheets reference](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html).

```python
plt.style.use("ggplot")
```

Ta cũng có thể lưu các biểu đồ với nhiều định dạng khác nhau (PDF, PNG, JPEG, ...) bằng hàm `savefig`.

```python
plt.style.use("default")

x = np.arange(150)
y = np.random.choice([-1, 1], x.size)
z = np.cumsum(y)
plt.plot(x, z)
plt.title("A random walk")
plt.savefig("a_rand_walk.pdf") 
```

**Bài tập**. Vẽ đồ thị các hàm số sau trên khoảng $[-2, 2]$
  - $y = x$
  - $y = x^2$
  - $y = |x|$ 
  
trên cùng một hệ biểu đồ (hệ trục) hay riêng lẻ trên từng biểu đồ. Thử dùng các style khác nhau mà matplotlib hỗ trợ.

## <a name="bieu_do_duong"/>Biểu đồ đường

Dữ liệu lợi nhuận của một công ty

```python
months = ["Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
profits = [2.0, 2.1, 2.2, 2.1, 2.3, 2.4] # millions
```

Vẽ **biểu đồ đường** (line chart) mô tả tăng trưởng lợi nhuận bằng hàm [`plot`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
plt.style.use("seaborn-v0_8-whitegrid")

plt.plot(months, profits)
plt.show()
```

Thiết lập chi tiết các giá trị của biểu đồ (tham khảo [Pyplot tutorial](https://matplotlib.org/stable/tutorials/introductory/pyplot.html))

```python
plt.plot(months, profits, marker="s", linewidth=2, color="blue")

plt.title("Company Profit per Month")
plt.xlabel("Month")
plt.ylabel("Profit (millions of dollars)")

plt.show()
```

Nếu muốn thể hiện bản thân lợi nhuận (cùng với tăng trưởng) thì nên vẽ biểu đồ sau

```python
plt.plot(months, profits, marker="s", linewidth=2)

plt.title("Company Profit per Month")
plt.xlabel("Month")
plt.ylabel("Profit (millions of dollars)")
plt.ylim(0, 2.5)

plt.show()
```

*Chọn đúng loại biểu đồ!*

Xem thêm việc biểu đồ có thể trình bày sai lệch về dữ liệu: [Misleading graph](https://en.wikipedia.org/wiki/Misleading_graph).

## <a name="bieu_do_quat">Biểu đồ quạt

Dữ liệu doanh số mỗi thể loại game

```python
genres = ["Sports", "Strategy", "Action", "Shooter", "Other"]
units_solds = [27_500, 11_500, 6_000, 3_500, 1_500] # unit
```

Vẽ **biểu đồ quạt** (pie chart) bằng hàm [`pie`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html)

```python
plt.pie(units_solds, labels=genres)
plt.show()
```

Các thiết lập chi tiết (tham khảo [Basic pie chart](https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html))

```python
def group_value(pct):
    return "{0:,}".format(int(round(pct/100, 2) * sum(units_solds)))

plt.pie(units_solds, labels=genres, startangle=90, counterclock=False, 
        autopct=group_value, pctdistance=1.2, labeldistance=1.4)

plt.title("Units Sold per Genre", y=-0.1)

plt.show()
```

Biểu đồ quạt thể hiện tỉ lệ

```python
plt.pie(units_solds, labels=genres, startangle=90, counterclock=False, 
        autopct="%.0f%%", pctdistance=1.2, labeldistance=1.4)

plt.title("Units Sold per Genre", y=-0.1)

plt.show()
```

Dữ liệu [`Iris Data Set`](https://archive.ics.uci.edu/ml/datasets/Iris).

```python
iris_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
iris = pd.read_csv('iris.data', names=iris_names)
iris 
```

Biểu đồ quạt thể hiện số lượng các giống hoa (species) trong bộ dữ liệu `Iris`.

```python
species_counts = iris.species.value_counts()
species_counts
```

```python
plt.pie(species_counts.values, labels=species_counts.index)
plt.show()
```

Hoặc vẽ nhanh dùng pandas (pandas sẽ dùng matplotlib).

```python
iris.species.value_counts().plot.pie()
plt.show()
```
