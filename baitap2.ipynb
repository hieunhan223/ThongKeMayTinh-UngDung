**3.1. (0.5 đ)** Tạo bảng dữ liệu MONHOC cho biết thông tin các môn học như sau

MH04 | Đồ Họa            | 45

Trong đó

  - SoTiet là số tiết

```python
duLieu_MONHOC = {
    "MaMH": ["MH01", "MH02", "MH03", "MH04"],
    "TenMH": ["Cơ Sở Dữ Liệu", "Trí Tuệ Nhân Tạo", "Truyền Thông", "Đồ Họa"],
    "SoTiet": [60, 45, 30, 45]
}

bangDuLieu_MonHoc = pd.DataFrame(duLieu_MONHOC)

# Hiển thị bảng
print(bangDuLieu_MonHoc)
```

**3.2. (0.5 đ)** Tạo bảng dữ liệu SINHVIEN cho biết thông tin các sinh viên như sau

MaSV | TenSV          | GioiTinh | NgaySinh

SV02 | Trần Văn Ba    | Nam      | 24/12/1992
SV03 | Lê Văn Ca      | Nữ       | 21/02/1990
SV04 | Phạm Thị Em    | Nữ       | 12/08/1991

Trong đó

  - TenSV là họ và tên
  - GioiTinh là giới tính

```python
duLieu_SINHVIEN = {
    "MaSV": ["SV01", "SV02", "SV03", "SV04"],
    "TenSV": ["Nguyễn Thị An", "Trần Văn Ba", "Lê Văn Ca", "Phạm Thị Em"],
    "GioiTinh": ["Nữ", "Nam", "Nữ", "Nữ"],
    "NgaySinh":["23/02/1990","24/12/1992","21/02/1990","12/08/1991"] 
}

bangDuLieu_SinhVien = pd.DataFrame(duLieu_SINHVIEN)

# Hiển thị bảng
print(bangDuLieu_SinhVien)
```

**3.3. (0.5 đ)** Tạo bảng dữ liệu KETQUA cho biết điểm thi của sinh viên cho môn học như sau

MaSV | MaMH | Diem
-----|------|-----
SV01 | MH01	| 5

SV02 | MH04 | 8
SV03 | MH01 | 4
SV04 | MH01 | 6
SV04 | MH02 | 7
SV04 | MH03 | 5
SV04 | MH04 | 8

Trong đó:
  - MaSV là mã sinh viên
  - MaMH là mã môn học
  - Diem là điểm thi

```python
duLieu_KETQUA = {
    "MaSV": ["SV01", "SV01", "SV02", "SV02", "SV03", "SV04", "SV04", "SV04", "SV04"],
    "MaMH": ["MH01", "MH02","MH01","MH04","MH01","MH01","MH02","MH03","MH04"],
    "Diem": [5, 4, 6, 8,4 , 6, 7, 5, 8]
}

bangDuLieu_KETQUA = pd.DataFrame(duLieu_KETQUA)

# Hiển thị bảng
print(bangDuLieu_KETQUA)
```

**3.4. (2.5 đ)** Từ 3 bảng dữ liệu trên, thực hiện các truy vấn để trả lời các yêu cầu sau (có thể viết thêm các hàm phụ nếu cần)

Câu 1: Đưa ra danh sách các môn học có số tiết không quá 45.

```python
cau1 = bangDuLieu_MonHoc[bangDuLieu_MonHoc["SoTiet"]<=45]
print(cau1)
```

Câu 2: Đưa ra danh sách các sinh viên nữ có năm sinh từ 1990 về trước.

```python
bangDuLieu_SinhVien["NgaySinh"] = pd.to_datetime(bangDuLieu_SinhVien["NgaySinh"], format="%d/%m/%Y")
cau2 = bangDuLieu_SinhVien[(bangDuLieu_SinhVien["GioiTinh"]=="Nữ") & (bangDuLieu_SinhVien["NgaySinh"]<"01/01/1990")]
print(cau2)
```

Câu 3: Đưa ra danh sách các môn học có số tiết nhiều nhất.

```python
soTietMax = bangDuLieu_MonHoc["SoTiet"].max()
cau3 = bangDuLieu_MonHoc[bangDuLieu_MonHoc["SoTiet"]== soTietMax]
print(cau3)
```

Câu 4: Cho biết tỉ lệ đậu/rớt tính trên tất cả các môn (từ 5 điểm trở lên là đậu).

```python
dau = (bangDuLieu_KETQUA["Diem"] >= 5).sum()

tong_so_mon = len(bangDuLieu_KETQUA)
ty_le_dau = (dau / tong_so_mon) * 100
ty_le_rot = 100 - ty_le_dau

print(f"Tỷ lệ đậu: {ty_le_dau:.2f}%")
print(f"Tỷ lệ rớt: {ty_le_rot:.2f}%")
```

Câu 5: Cho biết tên môn học có tỉ lệ rớt nhiều nhất.

```python
bangDuLieu_MonHoc_SoLuongSV = bangDuLieu_KETQUA.groupby("MaMH")["MaSV"].count()
bangDuLieu_MonHoc_Rot = bangDuLieu_KETQUA[bangDuLieu_KETQUA["Diem"] < 5].groupby("MaMH")["MaSV"].count()
bangDuLieu_ty_le_rot = (bangDuLieu_MonHoc_Rot / bangDuLieu_MonHoc_SoLuongSV).fillna(0).reset_index().rename(columns={"MaSV":"TyLeRot"})
tyLeRotMax = bangDuLieu_ty_le_rot["TyLeRot"].max()
cau5 = bangDuLieu_ty_le_rot[bangDuLieu_ty_le_rot["TyLeRot"]==tyLeRotMax]
print(cau5)
```

Câu 6: Cho biết điểm trung bình của sinh viên có tên là Phạm Thị Em.

```python
diem_tb_sinh_vien = bangDuLieu_KETQUA.groupby("MaSV")["Diem"].agg(DiemTrungBinh="mean").reset_index()
print(diem_tb_sinh_vien)
bangDuLieu_SinhVien_KetQua = pd.merge(bangDuLieu_SinhVien,diem_tb_sinh_vien,on="MaSV",how="left")
print(bangDuLieu_SinhVien_KetQua)
cau6 = bangDuLieu_SinhVien_KetQua[bangDuLieu_SinhVien_KetQua["TenSV"] =="Phạm Thị Em"]
print(cau6)
```

Câu 7: Đưa ra tên của các sinh viên không rớt môn nào mình học.

```python
sinh_vien_rot = bangDuLieu_KETQUA[bangDuLieu_KETQUA["Diem"] < 5]["MaSV"].unique()
sinh_vien_khong_rot = bangDuLieu_SinhVien[~bangDuLieu_SinhVien["MaSV"].isin(sinh_vien_rot)]
print(sinh_vien_khong_rot)
```
